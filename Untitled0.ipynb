```python
import os

# Updated paths based on your actual directory structure
train_path = "/content/train"
test_path  = "/content/test"

print("\n--- Exploratory Data Analysis ---")

try:
    # Get counts for training data
    hot_dog_count     = len(os.listdir(os.path.join(train_path, "hot_dog")))
    not_hot_dog_count = len(os.listdir(os.path.join(train_path, "not_hot_dog")))
    print(f"Hot dog images (train):     {hot_dog_count}")
    print(f"Not hot dog images (train): {not_hot_dog_count}")

    # Get counts for test data
    test_hot_dog_count     = len(os.listdir(os.path.join(test_path, "hot_dog")))
    test_not_hot_dog_count = len(os.listdir(os.path.join(test_path, "not_hot_dog")))
    print(f"Hot dog images (test):      {test_hot_dog_count}")
    print(f"Not hot dog images (test):  {test_not_hot_dog_count}")

except FileNotFoundError as e:
    print(f"❌ Error: Could not find the subfolders. {e}")
```

```python
IMG_HEIGHT = 64
IMG_WIDTH  = 64
BATCH_SIZE = 32

print("\n--- Preparing Data Generators ---")
# Using the augmented data generator for best results
augmented_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    validation_split=0.2
)

# Standard generator for validation (we don't augment validation data)
val_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = augmented_datagen.flow_from_directory(
    train_path,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)

validation_generator = val_datagen.flow_from_directory(
    train_path,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)
print("Class indices:", train_generator.class_indices)
```

```python
import matplotlib.pyplot as plt

# 1. Build a simple ANN
ann_model = keras.Sequential([
    # Flatten the 2D image into 1D (e.g., 128x128x3 becomes 49,152 pixels)
    layers.Flatten(input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),

    # Dense layers - Fixed curly quotes here
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid') # Output: probability of "hot dog"
])

# 2. Compile the model - Fixed curly quotes here
ann_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# 3. Show model summary
ann_model.summary()

# 4. Train for 10 epochs
history_ann = ann_model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=10
)

# 5. YOUR CODE HERE: Plot training vs validation accuracy
plt.figure(figsize=(10, 6))

# Use the 'history' dictionary to access the recorded metrics
plt.plot(history_ann.history['accuracy'], label='Training Accuracy')
plt.plot(history_ann.history['val_accuracy'], label='Validation Accuracy')

plt.title('ANN Training vs Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("\n--- Building and Training the CNN ---")
final_model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])

final_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = final_model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=15
)

```
